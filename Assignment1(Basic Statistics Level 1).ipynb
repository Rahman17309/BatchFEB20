## Question 16

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace, kurtosis
```

```python
data = laplace.rvs(loc=0, scale=1, size=1000)
```

```python
plt.hist(data, bins=30, density=True, alpha=0.5, color='b', label='Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Data with Positive Kurtosis')
plt.legend()

kurtosis_value = np.round(kurtosis(data), 2)
plt.text(2, 0.35, f'Kurtosis: {kurtosis_value}', fontsize=12, color='r')

plt.show()
```

```python
prob_MPG_greater_than_20 = np.round(1-stats.norm.cdf(20, loc = df20.MPG.mean(), scale = df20.MPG.std()),3)
print('p(MPG>20)=',(prob_MPG_greater_than_20))
```

```python
prob_MPG_less_than_50 = np.round(stats.norm.cdf(50, loc = df20.MPG.mean(), scale = df20.MPG.std()),3)
print('P(MPG<50)=',(prob_MPG_less_than_50))
```

```python
prob_MPG_greaterthan20_and_lessthan50= (prob_MPG_less_than_50) - (prob_MPG_greater_than_20)
print('P(20<MPG<50)=',(prob_MPG_greaterthan20_and_lessthan50))
```

```python
mean, cov = [0, 0], [(1, .6), (.6, 1)]
x, y = np.random.multivariate_normal(mean, cov, 100).T
y += x + 1

f, ax = plt.subplots(figsize=(6, 6))

ax.scatter(x, y, c=".3")
ax.set(xlim=(-3, 3), ylim=(-3, 3))

# Plot your initial diagonal line based on the starting
# xlims and ylims.
diag_line, = ax.plot(ax.get_xlim(), ax.get_ylim(), ls="--", c=".3")

def on_change(axes):
    # When this function is called it checks the current
    # values of xlim and ylim and modifies diag_line
    # accordingly.
    x_lims = ax.get_xlim()
    y_lims = ax.get_ylim()
    diag_line.set_data(x_lims, y_lims)

# Connect two callbacks to your axis instance.
# These will call the function "on_change" whenever
# xlim or ylim is changed.
ax.callbacks.connect('xlim_changed', on_change)
ax.callbacks.connect('ylim_changed', on_change)

plt.show()
```
